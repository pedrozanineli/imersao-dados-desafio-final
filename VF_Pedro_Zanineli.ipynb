##**1. Introdução ao tema de *drug discovery*, histórico e problemática**

Estas evidências, assim como as encontradas em Ohallo II, demonstram a longevidade da busca por substâncias para uso medicinal.

  De maneira mais distante, hoje continuamos a buscar - com uma complexidade muito maior - formas de inibir os efeitos causados por patologias e tratar de problemas enfrentados diariamente, como evidenciado pela pandemia causada pelo vírus Sars-Cov-2, em que cientistas de diversos países voltaram-se para o desenvolvimento de um tratamento medicamentoso. ###**Aplicação da ótica estatística**

Apoia-se então em medidas de tendência central (como é o caso da média), medidas de dispersão (tal como o desvio-padrão), tabelas (como a distribuição de frequências) e gráficos (como é o caso do histograma).

Apesar de a **média** não apresentar fielmente a distribuição de um conjunto de elementos, o quociente resultante da soma e da quantidade de dados pode servir de apoio para termos uma ideia de como os dados comportam-se.

Fórmula da média: $\bar{x} =\frac{\sum_{i=1}^{k}x_i*f_i}{n}$

```python
# Célula-21

exp.mean()
```

Note que na saída do código abaixo, ainda é mostrado o **count** (a contagem dos valores), a **std** - standard deviation, ou desvio padrão, calculado pela raiz quadrada da variância (soma dos quadrados do desvio) - e os **quartis** - divisão do conjunto em quatro partes iguais, separando os 25% valores menores dos 75% valores maiores, calculado por $Q_j=x_k+(\frac{j(n+1)}{4}-k)(x_{k+1}-x_k)$.

O desvio padrão tem grande importância para entendermos a dispersão do conjunto, ao passo de que indica numericamente o quanto um conjunto está disperso - ou seja, quanto menor for o seu valor, menos disperso está o conjunto.


```python
# Célula-22
# Usamos a expressão loc para definir um intervalo de análise

exp.loc[:,'g0':'g50'].describe()
```

Como mencionado na "Aplicação da ótica estatística", a plotagem de gráficos como o histograma - retângulos justapostos onde a base representa o intervalo da classe ao passo de que a altura representa a frequência - é comum nesse campo de análise, servindo de grande utilidade para analisarmos os elementos.

É importante observar que na plotagem desse histograma é possível encontrar uma simetria entre os lados que o compõem, mas que não garantem que a distribuição aconteceu assim.

Segundo a mestra em genética, conservação e biologia evolutiva Vanessa Leiko, essa normalização é uma questão bastante comum para a área, não permitindo utilizar unicamente o histograma para concluir como os valores estão distribuídos.

A vantagem em encontrar-se uma curva normal é por conseguir encontrar padrões entre as amostras - assim como feito por Frederick Gauss, por volta do século XIX. Assim, indicar um comportamento que variam em torno de um valor médio pode sustentar uma hipótese ou indicar certidão no caminho a ser seguido [[13]](https://www.voitto.com.br/blog/artigo/distribuicao-normal).

Ainda assim, é possível notar que, se diminuirmos o intervalo das classes, teremos uma visão ainda mais parecida a de uma distribuição normalizada - assim como feito na célula 24, triplicando a quantidade de intervalos. Na célula 25, os gráficos são colocados lado a lado para melhor observação.

###**Agregando funções na análise**

Mais uma forma de criarmos análises entre os dados expostos é agregando funções por meio de um crosstab. Por exemplo, se quisermos comparar as médias de um grupo de valores, podemos utilizar o código indicado na célula 27, comparando dose, tempo e tamanho do composto g0.

```python
# Célula-27

pd.crosstab([exp['dose'],exp['tempo']],exp['tratamento'], values=exp['g0'], aggfunc='mean')
```

Com base na tabela acima, podemos levantar uma reflexão que se faz de grande importância que é o tempo de effetividade de um medicamento. Caso a droga apresente efeito depois de, por exemplo, 72 horas, pode na verdade não ser mais algo positivo, mas um **efeito colateral** do medicamento.

No caso apresentado, vale utilizar a variância para indicar o quão longe os dados estão posicionados entre as amostras, na tentativa de buscar uma possível correlação entre eles.

####**Utilização do modelo de regressão logística**

Usado para medir a probabilidade de uma certa classe ou evento existente passível de determinar duas condições opostas, o modelo de classificação de regressão analisa diversas componentes, atribuindo valores de 0 e 1, somando os resultados positivos para determinação [[16]](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html).

Apesar de já termos realizado a importação de algumas bibliotecas, será necessário agora utilizar novas que não foram previamente aplicadas. Todas fazem parte do *Sci-kit* learning [[17]](https://scikit-learn.org/stable/user_guide.html), biblioteca com ferramentas consideradas simples e eficientes para predição de análise de dados.

```python
# Célula-43

# Importação do modelo de regressão logística
from sklearn.linear_model import LogisticRegression

# Importação do modelo de teste
from sklearn.model_selection import train_test_split
```

Para construção do modelo, é necessário com que parte dos dados seja destinado para o treino, aprendizado do modelo, e outra parte para o teste em si, aplicação prática do que foi treinado.

Essa proporção é uma grande variável, mas tende a funcionar em uma proporção de 20% para treino e 80% para testes. Configura-se como algo de bastante atenção, ao passo de que, caso seja treinado com um volume grande de informações, um *overfit* (treino demasiado) pode acontecer.

Se um *overfit* acontecesse, seria de muita facilidade fazer com que o modelo acertasse os resultados, afinal, poderia nem mesmo acertar, mas inferir, já saber a resposta dos mesmos.

É importante ressaltar que no código de teste e treino da célula 44, é passado dois parâmetros não obrigatórios, o stratify e o random state. A utilização dessas orientações ao comando tem como objetivo construir uma predição mais fidedigna possível, que não tenha tantas alterações.

```python
# Célula-44

# Passando a assinatura celular
x = combinados.select_dtypes('float64')

# Passando os dados de ativação de MoA
y = combinados['ativo_moa']

# Alocando os valores encontrados para teste e treino em variáveis
x_treino, x_teste, y_treino, y_teste = train_test_split(x,y, test_size=0.2, stratify = y, random_state=400)

# Atribuir a uma variável o modelo de regressão logística
# Note que, quando o código roda com a linha abaixo, aparece um warning
# modelo_rlogistica = LogisticRegression()

# Para resolver isso, usar o parâmetro max_iter
# Entendido como o máximo número de interações dadas para conversão pelos solucionadores
modelo_rlogistica = LogisticRegression(max_iter=1000)

# Nosso treino precisa receber também as variáveis de teste
train_test_split(x,y)

# Agora preciso treinar o modelo - igual os parâmetros de uma função, que são as nossas labels, as assinaturas celulares
modelo_rlogistica.fit(x_treino,y_treino)

# Agora preciso verificar como o meu modelo está se comportando - verificar a acurácia do modelo
modelo_rlogistica.score(x_teste,y_teste)
```

Note que a acurácia encontrada para o nosso modelo de predição foi de praticamente **63%**. Temos que refletir agora se, construído dessa forma e com esse resultado, será suficiente.

É necessário levar em consideração que os medicamentos testados clinicamente devem ter um bom índice de certidão, a fim de que seja possível levar os compostos que apresentaram uma possível funcionalidade para seguintes análises e evitar com que compostos que não detenham aplicabilidade tenham seus testes terminados.

Pode-se entender então que o modelo ainda não é suficiente, por se tratar de um caso em que se é necessário ter uma taxa efetiva superior.

####**Comparação superficial do valor encontrado**

Podemos utilizar um método ainda do *Sci-kit learn* para formularmos um outro modelo, com regras simples e para efeito de comparação de outros protótipos. Usaremos, então, o *Dummy Classifier*, uma classificação fictícia.

```python
# Célula-45

# Fazendo uma comparação com o próprio banco do sci-kit
from sklearn.dummy import DummyClassifier

# Método para calcular a acurácia
from sklearn.metrics import accuracy_score

modelo_dummy = DummyClassifier('most_frequent')
modelo_dummy.fit(x_treino,y_treino)

# Fazer uma predição dos nossos testes
previsao_dummy = modelo_dummy.predict(x_teste)

# Para construir a acurácia
accuracy_score(y_teste,previsao_dummy)
```

####**Utilização do modelo de árvore de decisões**

Uma árvore de decisões pode ser entendido como o mapeamento dos possíveis resultados com base em uma série de escolhas relacionadas [[18]](https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html).

Pode ser entendido como uma árvore por sempre dividir o conjunto em dois, buscando uma homogeinidade entre cada um dos grupos.

```python
# Célula-47

# Importação do modelo
from sklearn.tree import DecisionTreeClassifier

x = combinados.select_dtypes('float64')
y = combinados['ativo_moa']

x_treino, x_teste, y_treino, y_teste = train_test_split(x,y, test_size=0.2, stratify = y, random_state=400)

# Aplicando o modelo da árvore
# max_depth entendido como o número de ramificações, galhos, a serem feitos
modelo_arvore = DecisionTreeClassifier(max_depth=3)
modelo_arvore.fit(x_treino,y_treino)
modelo_arvore.score(x_teste,y_teste)
```

Podemos considerar que a acurácia desse modelo foi de **61%**, inferior ao que encontrado na última análise. Porém, pode ser interessante o aumento do número de galhos, isto é, das divisões e então das escolhas, buscando uma análise mais bem detalhada.

Na célula 48, é plotado a árvore decisões executada na 47, permitindo visualização do processo feito. Note:

1. Índice de **gini**: o quão homogêneo o conjunto é;
2. **samples**: número de amostras;
3. **value**: o intervalo de valores;
4. **class**: a classificação do conjunto, se foi ou não ativado.

```python
# Célula-48

from sklearn import tree
fig, ax = plt.subplots(figsize=(15,10), facecolor='k')
tree.plot_tree(modelo_arvore,
               ax=ax,
               fontsize=16,
               rounded=True,
               filled=True,
               feature_names = x_treino.columns,
               class_names=['Não Ativado', 'Ativado'])

plt.show()
```

####**Aumentando a quantidade de decisões da árvore**

Como mencionado anteriormente, uma forma de melhorar o valor encontrado é de aumentar a quantidade de divisões. O método "*for*", laço de repetição comunmente utilizado na programação [[19]](https://www.devmedia.com.br/conceitos-e-exemplos-do-for-lacos-de-repeticoes-estrutura-da-linguagem-parte-2/18871), irá auxiliar na análise dos resultados, já que será passado como parâmetro do max_depth um valor auto incrementado e sempre armazenar a acurácia em uma lista.

```python
# Célula-49

# Também quero fazer testes para fazer análises posteriormente
teste=[]
treino=[]

# Criar um for para fazer o processo diversas vezes
for i in range(1,15):
  modelo_arvore = DecisionTreeClassifier(max_depth=i)
  modelo_arvore.fit(x_treino,y_treino)
  teste.append(modelo_arvore.score(x_teste,y_teste))
  treino.append(modelo_arvore.score(x_treino,y_treino))
```

```python
# Célula-51

teste
```

Observe os valores de acurácia encontrados para os testes e treinos demonstram um problema. Apesar de conseguirmos chegar em uma acurácia de treino de **80%**, diminuímos nossa acurácia de teste para **58%**.

Como citado anteriormente, o processo do *overfit* aconteceu, em que passou, à medida que passou a ter uma certidão tão notável, acabou por diminuir os valores dos testes em si. A análise pode ser visualmente feita na célula 52.

Por conta de não ser desejado, ou é necessária a diminuição do max_depth ou a tentativa de uma outra forma de análise, criação de uma árvore de decisão que utiliza amostras aleatórias, por exemplo.

```python
# Célula-52

plt.figure(figsize=(12,10))

sns.lineplot(x=range(1,15), y=teste, label='teste', palette='pastel')
sns.lineplot(x=range(1,15), y=treino, label='treino', palette='pastel')

plt.title('Relação entre os valores de teste e treino', fontsize=20)
plt.xlabel('Quantidade de treinos')
plt.show()
```

###**Utilização do modelo *random forest***

Considerado ainda um modelo relativamente simples, treina um número de decisões em uma árvore em diferente sub-amostras do banco de dados e usar o cálculo da média para aumentar a acurácia e o controle de *overfit* [[20]](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html).

```python
# Célula-53

from sklearn.ensemble import RandomForestClassifier

# Selecionar também outros elementos
x = combinados.drop(['id','n_moa','ativo_moa','composto'], axis=1)

# Não colocar o "composto" junto por justamente querer generalizar, é aquele que quer chegar a conclusão
# Pode "vazar dados", ter uma variável que já é a resposta

# Vai receber o x
# Como está em strings, vai dar erro - para resolver: pd.dummies
x = pd.get_dummies(x, columns=['tratamento','dose','tempo'])

y = combinados['ativo_moa']

x_treino, x_teste, y_treino, y_teste = train_test_split(x,y, test_size=0.2, stratify = y, random_state=376)

modelo_randomforest = RandomForestClassifier()

modelo_randomforest.fit(x_treino,y_treino)
modelo_randomforest.score(x_teste,y_teste)
```

Apesar de ainda ser baixa, a acurácia apresentada de **68,5%** foi maior do que os testes realizados de outras formas com as árvores de decisão.

O objetivo dessa seção final é de trazer uma conclusão da problemática apresentada, de forma a indicar o que podemos concluir com base no exposto e quais os próximos passos a serem tomados.

Após entendimento dos dados coletados, com a análise das variáveis dos compostos gênicos, viabilidade celular, mecanismos de ativação, assinatura celular e seus possíveis valores, buscando relações entre os objetos de análise, como o tipo de tratamento, dose e tempo, apoiando-se na ciência de dados com cálculos estatísticos e plotagens de gráficos, foi então possível aplicar o *Machine Learning*.

Ao refletirmos a respeito dos valores de acurácia apresentados na última divisão, podemos compreender que ainda são insuficientes para a aplicação farmacêutica. Outros modelos podem ser aplicados em busca de resultados melhores, mas que não foi possível no presente trabalho por conta do tempo estipulado para sua realização.

Não apenas um outro modelo, como também a mudança do questionamento inicial. Isto é, se gostariamos de saber se os mecanismos serão ativados, um possível caminho a ser seguido é se alguum dos mecanismos será ativado, não pensando como um todo, mas de maneira individual e que possa trazer resultados palpáveis.

Mesmo que os valores de acurácia não sejam tão sginificativos, é importante compreender que ainda faz-se uma tarefa desafiadora encontrar o modelo de análise mais coerente, levando em conta que deve ser aplicável e que traga de fato mudanças para a sociedade. Cada vez mais, estamos no caminho para uma medicina mais humanizada e individualizada.

Um exemplo disso é o caso da primeira droga customizada para uma pessoaapenas, a *Milasen* [[21]](https://www.milasmiracle.org/milasen). A menina Mila, depois de já ter tido grande dificuldade na vida, ganhou esperança com a criação de um medicamento próprio para ela. Por fim, vale ressaltar que essa esperança não se restringe à Mila, mas a todos, com a esperança na ciência, na medicina e a busca por um mundo melhor e mais igualitário.
